```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as ptl
```

```python
dataset=pd.read_excel('set_de_datos_con_perspectiva_de_genero.xlsx')
```

### Limpiar Dataset

```python
dataset.head()
```

```python
c_dataset=dataset.loc[:, ['VIOLENCIA_DE_GENERO', 'V_FISICA', 'V_PSIC', 'V_ECON', 'V_SEX', 'V_SOC', 'V_AMB', 'V_SIMB', 'GENERO_ACUSADO/A', 'GENERO_DENUNCIANTE', 'LINK']]
c_dataset=c_dataset.rename(columns={'GENERO_ACUSADO/A': 'GENERO_ACUSADO'})
```

```python
c_dataset.dropna(inplace=True)
```

```python
c_dataset.head()
```

```python
c_dataset.isna().sum()
```

```python
c_dataset.GENERO_ACUSADO.unique()
```

```python
c_dataset.GENERO_DENUNCIANTE.unique()
```

```python
c_dataset.loc[~c_dataset.GENERO_ACUSADO.isin(['femenino', 'masculino']), 'GENERO_ACUSADO']='otro'
c_dataset['GENERO_DENUNCIANTE'] = dataset['GENERO_DENUNCIANTE'].replace('Femenino' ,'femenino')
for i in range(1,8):
    c_dataset.iloc[:,i]=c_dataset.iloc[:,i].replace('no_corresponde', 'no')
```

```python
c_dataset.drop_duplicates(inplace = True) 
```

```python
col=c_dataset.columns
for i in range(0, 10):  
    s_col=col[i]
    c_dataset[s_col]=c_dataset[s_col].replace('s/d', 'no')
```

```python
violencia=c_dataset[c_dataset['VIOLENCIA_DE_GENERO']=='si']
no_violencia=c_dataset[c_dataset['VIOLENCIA_DE_GENERO']=='no']
print('Casos de violencia: ', len(violencia))
print('Casos de no violencia: ',len(no_violencia))
```

### Procesar el contenido del link

```python
import obtener_contenido_archivo as oca
```

```python
#lista=oca.cargar_desde_carpeta(violencia['LINK'])
lista=oca.descargar_de_drive(violencia['LINK'])
```

```python
#lista2=oca.cargar_desde_carpeta(no_violencia.iloc[0:800,10])
lista2=oca.descargar_de_drive(no_violencia['LINK'])
```

```python
violencia['TXT_DECLARACION']=lista
no_violencia=no_violencia.iloc[0:800,:]
no_violencia['TXT_DECLARACION']=lista2
```

```python
violencia.drop( violencia[ violencia['TXT_DECLARACION']=='Archivo no encontrado'].index , inplace=True)
violencia.drop( violencia[ violencia['TXT_DECLARACION']=='NADA'].index , inplace=True)
no_violencia.drop( no_violencia[ no_violencia['TXT_DECLARACION']=='Archivo no encontrado'].index , inplace=True)
no_violencia.drop( no_violencia[ no_violencia['TXT_DECLARACION']=='NADA'].index , inplace=True)
```

```python
print(len(violencia))
print(len(no_violencia))
```

```python
violencia.dropna(inplace=True)
no_violencia.dropna(inplace=True)
```

```python
unificado=pd.concat([violencia, no_violencia])
```

```python
import mod_pred_violencia as ent
```

```python
corpus=ent.spacy_steaming(unificado['TXT_DECLARACION'], ['ADJ', 'VERB', 'NOUN'])
```

```python
v_corpus_spacy, vectorizer=ent.tf_itf(corpus, 5000)
```

```python
import pickle
pickle.dump(vectorizer,open("countVectorizer_ntkl","wb"))
```

```python
col=unificado.columns
for i in range(0, 10):  
    s_col=col[i]
    unificado= pd.get_dummies(unificado, columns=[s_col])
```

```python
for item in unificado.columns.tolist():
    if "_no" in str(item):
        unificado.drop([item], axis='columns', inplace=True)
```

```python
y=unificado.iloc[:,2:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(v_corpus_spacy, y, test_size=0.10, random_state=1000)
```

```python
print(len(X_train))
print(len(X_test))
```

## Entrenar modelos

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

## Guardar modelos y performance

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV

# Main settings
epochs = 30
batch_size =6

# Parametros grid for grid search
param_grid = dict(nodes_1=[20, 60, 120],
                  nodes_2=[20, 60, 120],
                  drop_1= [0.2, 0.3, 0.4, 0.5],
                  drop_2= [0.2, 0.3, 0.4, 0.5])
    
model = KerasClassifier(build_fn=ent.create_model,
                            epochs=epochs, batch_size=batch_size,
                            verbose=False)
grid = RandomizedSearchCV(estimator=model, param_distributions=param_grid,
                              cv=4, verbose=1, n_iter=5)
grid_result = grid.fit(X_train, y_train)

# Evaluate testing set
test_accuracy = grid.score(X_test, y_test)

print(grid_result.best_score_)
print(grid_result.best_params_)
print(test_accuracy)
 
```

```python
grid_result.best_params_
modelo_final=ent.create_model(60, 20, 0.4, 0.4)
modelo_final.fit(X_train, y_train.to_numpy() , batch_size = 6, epochs = 30)
```

```python
y_pred=modelo_final.predict(X_test)
```

```python
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred>0.5)
```

```python
multilabel_confusion_matrix(y_test, y_pred>0.5)
```
